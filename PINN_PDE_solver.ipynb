```python
## WIP: construct CNN
# here's an example of a neural network with 2 inputs (x & t) and 1 output

class NeuralNet(nn.module):
    def __init__(self, layers):
        super().__init__()

        self.depth = len(layers) - 1

        self.activation = nn.Tanh # change activation func here

        layer_list = list()
        for i in range(self.depth - 1):
            layer_list.append( ('layer_%d' % i, nn.Linear(layers[i], layers[i+1]))) 
            # create new layers each called layer_n with n being the number of layer
            layer_list.append(('activation_%d' % i, self.activation()))
            # also add activation

        # add the final layer here
        layer_list.append(('layer_%d' % (self.depth - 1), nn.Linear(layers[-2], layers[-1])))

        layer_dict = OrderedDict(layer_list)
        
        self.layers = nn.Sequential(layer_dict)

    def forward(self, x):
        out = self.layers(x)
        return out
```

```python
## WIP: set up parameters and optimizer

iter = 100 # set number of iterations here.
batch_size = 50
learning_rate = 0.1 # set learning rate here

layers = [1, 5, 5, 1] # just set an arbitrary amount of hidden layers here. can play around with it later
nn_model = NeuralNet(layers)
optimizer = optim.Adam(nn_model.parameters(), lr = learning_rate)

# WIP: training loop

```
